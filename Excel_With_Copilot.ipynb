#Excel With Copilot


1.Use Copilot to identify and summarize missing values in the dataset.
- Steps to Identify & Summarize Missing Values Using Copilot

Open your dataset

Open the Excel file containing your dataset.

Make sure the data is formatted as a Table (Ctrl + T).

Open Copilot

Click on the Copilot icon in Excel

Ask Copilot to find missing values

Type a prompt such as:

“Identify missing values in this dataset and summarize them by column.”

Copilot Analysis

Copilot scans each column and:

Detects blank cells

Counts missing values per column

Identifies patterns (e.g., which columns have the most missing data)

Summary Output

Copilot provides:

A summary table showing:

Column Name

Number of Missing Values

Percentage of Missing Values

A short text explanation, for example:

“The Profit and Discount columns contain missing values, while Order ID and Region are complete.”

Example
Column Name	 Missing Values	Percentage
Sales	         5 	             2%
Profit	       12	             6%
Discount	     8	             4%
Region	       0	             0%


2. Use Copilot to fill missing values in the Region and category columns.
-  Using Copilot to Fill Missing Values in Region and Category
Steps:

Open the Dataset

Open your Excel file.

Convert the data into a Table (Ctrl + T) so Copilot can analyze it properly.

Launch Copilot

Click the Copilot icon from the Excel toolbar.

Give Copilot the Prompt

Enter a command such as:

“Fill missing values in the Region and Category columns using the most frequent values.”

(we can also say “based on similar records” if your dataset contains patterns.)

Copilot Processing

Copilot:

Scans existing values in Region and Category

Identifies the most common or logically matching entries

Automatically fills blank cells

Review Suggested Changes

Copilot highlights the updated cells.

Review and confirm the changes.

Example:

Before:

Order ID	Region   	Category
ORD101	  North	    Furniture
ORD102		   _       Furniture
ORD103	  South	        _

After Copilot:

Order ID	Region	 Category
ORD101	North	     Furniture
ORD102	North	     Furniture
ORD103	South	     Furniture

4. Create a new calculated column Total Revenue.
- Formula for Total Revenue = Units Sold × Unit Price
Units Sold → Column H

Unit Price → Column I

New column Total Revenue → Column J

Step 1: Add column header

In cell J1, type: Total Revenue

Step 2: Enter formula

In J2, enter: =H2*I2

Step 3: Apply to all rows

Press Enter

Double-click the fill handle (small square at bottom-right of J2)

The formula applies to all rows

Total Revenue is now calculated for every order.

8. Build a dashboard combining key KPIs and charts.
- Step 1: Open Copilot

Open Excel file

Click Copilot

Step 2: Create KPI Cards

Type this prompt:

“Create KPI cards for Total Sales, Total Profit, Average Profit, and Total Units Sold.”

Click Apply

Copilot will generate KPI visuals automatically.

Step 3: Add Charts

Enter prompts one by one:

a. Average Profit by Region

“Create a column chart showing average profit by region.”

b. Sales by Category & Region

“Create a PivotChart showing total sales by category and region.”

c. Monthly Sales Trend

“Create a line chart showing monthly total sales trend.”

Click Apply after each.

Step 4: Add Filters (Slicers)

Type:

“Add slicers for Region and Category to filter all charts.”

Step 5: Arrange as Dashboard

Type:

“Arrange all KPIs and charts into a single dashboard layout.”

 B. Manual Dashboard Structure
Top Row → KPI Cards

Use formulas:

Total Sales = SUM(Total Revenue)
Total Profit = SUM(Profit)
Average Profit = AVERAGE(Profit)
Total Units Sold = SUM(Units Sold)


Format as:

Bold

Large font

Border / Shape (Insert → Shapes)

Middle Section → Charts

Column Chart → Avg Profit by Region

PivotChart → Sales by Category & Region

Bottom Section → Trend & Filters

Line Chart → Monthly Sales

Slicers → Region, Category

- “A dashboard was built in Excel combining KPI cards, PivotCharts, and slicers to provide an interactive summary of sales and profit performance across regions and categories, supporting data-driven decision making.”
